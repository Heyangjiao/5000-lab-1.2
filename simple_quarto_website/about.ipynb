```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(x, y, c='blue', label='Random Data', alpha=0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter')
plt.legend()
plt.grid(True)

# Display the plot in the Jupyter Notebook
plt.show()
```
